## The Commute

The daycare I will choose should be either close enough to my home or work. I have already very limited time to spend with my daughter, so I don’t want to waste time on the road. Minimizing the commute will be better for me and also my daughter, here are the locations of the possible options

![location](img/daycare_locations.png)


I live Downtown so the daycares that are located in Downtown will be walking distance for me and since I will be at UBC for all day, the daycare at UBC also be in walking distance. Shaugnessy and Montesorri Day Care Society are located on my way to work. Let's take a look at how much time will spend:


```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

names_dayc = ("VSOCC", "YMCA","UBC", "Shaughnessy", "Montessori")
mins = [10,15, 20, 15, 12]

plt.bar(names_dayc, mins)
plt.xlabel('Daycares')
plt.ylabel('Commute(in mins)')

plt.show()
```
